## 1. More on SVD

## a)

To show $\mathbf{A} = \mathbf{U}\mathbf{S}\mathbf{V^T} = \sum_\alpha s_\alpha \mathbf{u_\alpha} \mathbf{v_\alpha^T}$. Let $\mathbf{A}$ be an $mxn$ array. $\mathbf{U}$, $\mathbf{S}$, and $\mathbf{V}$ are then of the form:

\begin{array}{c|c}
\mathbf{U} = 
\begin{bmatrix}
    u_{11} & u_{12} & \ldots & u_{1r} \\
    u_{21} & u_{22} & \ldots & u_{2r} \\
    \vdots & \vdots & \ddots & \vdots \\
    u_{m1} & u_{m2} & \ldots & u_{mr}
\end{bmatrix} \hspace{1cm} \mathbf{S} = 
\begin{bmatrix}
    s_{1} & 0 & \ldots & 0 \\
    0 & s_2 & \ldots & 0 \\
    \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & \ldots & s_r
\end{bmatrix}\hspace{1cm}\mathbf{V} = 
\begin{bmatrix}
    v_{11} & v_{12} & \ldots & v_{1r} \\
    v_{21} & v_{22} & \ldots & v_{2r} \\
    \vdots & \vdots & \ddots & \vdots \\
    v_{n1} & v_{n2} & \ldots & v_{nr}
\end{bmatrix}
\end{array}

Consider first $\mathbf{U}\mathbf{S}$:
\begin{array}{c|c}
\mathbf{U}\mathbf{S} = 
\begin{bmatrix}
    u_{11} & u_{12} & \ldots & u_{1r} \\
    u_{21} & u_{22} & \ldots & u_{2r} \\
    \vdots & \vdots & \ddots & \vdots \\
    u_{m1} & u_{m2} & \ldots & u_{mr}
\end{bmatrix}
\begin{bmatrix}
    s_{1} & 0 & \ldots & 0 \\
    0 & s_2 & \ldots & 0 \\
    \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & \ldots & s_r
\end{bmatrix} = 
\begin{bmatrix}
    s_1u_{11} & s_2u_{12} & \ldots & s_ru_{1r} \\
    s_1u_{21} & s_2u_{22} & \ldots & s_ru_{2r} \\
    \vdots & \vdots & \ddots & \vdots \\
    s_1u_{m1} & s_2u_{m2} & \ldots & s_ru_{mr}
    \end{bmatrix} 
\end{array}

Now consider the transpose of $\mathbf{V}$:


\begin{array}{c|c}
\mathbf{V^T}= 
\begin{bmatrix}
     v_{11} & v_{21} & \ldots & v_{n1} \\
    v_{12} & v_{22} & \ldots & v_{n2} \\
    \vdots & \vdots & \ddots & \vdots \\
    v_{1r} & v_{2r} & \ldots & v_{nr}
\end{bmatrix}
\end{array}

Therefore:
\begin{array}{c|c}
\mathbf{U}\mathbf{S}\mathbf{V^T} = 
\begin{bmatrix}
    s_1u_{11} & s_2u_{12} & \ldots & s_ru_{1r} \\
    s_1u_{21} & s_2u_{22} & \ldots & s_ru_{2r} \\
    \vdots & \vdots & \ddots & \vdots \\
    s_1u_{m1} & s_2u_{m2} & \ldots & s_ru_{mr}
\end{bmatrix} 
\begin{bmatrix}
     v_{11} & v_{21} & \ldots & v_{n1} \\
    v_{12} & v_{22} & \ldots & v_{n2} \\
    \vdots & \vdots & \ddots & \vdots \\
    v_{1r} & v_{2r} & \ldots & v_{nr}
\end{bmatrix}
\end{array}

\begin{array}{c|c}
\mathbf{U}\mathbf{S}\mathbf{V^T} = 
\begin{bmatrix}
    s_1u_{11}v_{11} + s_2u_{12}v_{12} + \ldots + s_ru_{1r}v_{1r} & s_1u_{11}v_{21} + s_2u_{12}v_{22} + \ldots + s_ru_{1r}v_{2r} & \ldots & s_1u_{11}v_{n1} + s_2u_{12}v_{n2} + \ldots + s_ru_{1r}v_{nr}\\
    s_1u_{21}v_{11} + s_2u_{22}v_{12} + \ldots + s_ru_{2r}v_{1r} & s_1u_{21}v_{21} + s_2u_{22}v_{22} + \ldots + s_ru_{2r}v_{2r} & \ldots & s_1u_{21}v_{n1} + s_2u_{22}v_{n2} + \ldots + s_ru_{2r}v_{nr}\\
    \vdots & \vdots & \ddots & \vdots \\
    s_1u_{m1}v_{11} + s_2u_{m2}v_{12} + \ldots + s_ru_{mr}v_{1r} & s_1u_{m1}v_{21} + s_2u_{m2}v_{22} + \ldots + s_ru_{mr}v_{2r} & \ldots & s_1u_{m1}v_{n1} + s_2u_{m2}v_{n2} + \ldots + s_ru_{mr}v_{nr}
\end{bmatrix}
\end{array}


\begin{array}{c|c}
\mathbf{U}\mathbf{S}\mathbf{V^T} = 
\begin{bmatrix}
    s_1u_{11}v_{11} & s_1u_{11}v_{21} & \ldots & s_1u_{11}v_{n1}\\
    s_1u_{21}v_{11} & s_1u_{21}v_{21} & \ldots & s_1u_{21}v_{n1}\\
    \vdots & \vdots & \ddots & \vdots \\
    s_1u_{m1}v_{11} & s_1u_{m1}v_{21} & \ldots & s_1u_{m1}v_{n1}
\end{bmatrix} +
\begin{bmatrix}
    s_2u_{12}v_{12} & s_2u_{12}v_{22} & \ldots & s_2u_{12}v_{n2}\\
    s_2u_{22}v_{12} & s_2u_{22}v_{22} & \ldots & s_2u_{22}v_{n2}\\
    \vdots & \vdots & \ddots & \vdots \\
    s_2u_{m2}v_{12} & s_2u_{m2}v_{22} & \ldots & s_2u_{m2}v_{n2}
\end{bmatrix} + \ldots +
\begin{bmatrix}
    s_ru_{1r}v_{1r} & s_ru_{1r}v_{2r} & \ldots & s_ru_{1r}v_{nr}\\
    s_ru_{2r}v_{1r} & s_ru_{2r}v_{2r} & \ldots & s_ru_{2r}v_{nr}\\
    \vdots & \vdots & \ddots & \vdots \\
    s_ru_{mr}v_{1r} & s_ru_{mr}v_{2r} & \ldots & s_ru_{mr}v_{nr}
\end{bmatrix}
\end{array}

\begin{array}{c|c}
\mathbf{U}\mathbf{S}\mathbf{V^T} = s_1
\begin{bmatrix}
    \mathbf{u_{1}}v_{11} & \mathbf{u_{2}}v_{21} & \ldots &  \mathbf{u_{r}}v_{n1}\\
\end{bmatrix} + s_2
\begin{bmatrix}
    \mathbf{u_{2}}v_{12} & \mathbf{u_{2}}v_{22} & \ldots &  \mathbf{u_{r}}v_{n1}\\
\end{bmatrix} + \ldots + s_r
\begin{bmatrix}
    \mathbf{u_{r}}v_{1r} & \mathbf{u_{r}}v_{2r} & \ldots &  \mathbf{u_{r}}v_{nr}\\
\end{bmatrix}
\end{array}

$\mathbf{U}\mathbf{S}\mathbf{V^T} = s_1 \mathbf{u_{1}} \otimes \mathbf{v_{1}} +  s_2 \mathbf{u_{2}} \otimes \mathbf{v_{2}} + \ldots +  s_r \mathbf{u_{r}} \otimes \mathbf{v_{r}}$

$\mathbf{U}\mathbf{S}\mathbf{V^T} = \sum_\alpha s_\alpha \mathbf{u_\alpha} \mathbf{v_\alpha^T}$

## b)

```python
m = 10
n = 6
r = min(m, n)

A = np.random.rand(m, n)
```

```python
U, Sdiag, V_T = np.linalg.svd(A,0)

print(Sdiag)
```

```python
# plot
component_matrices = np.zeros(shape=(r, m, n))

for i in range(len(component_matrices)):
    u = U[:, i]
    v = V_T[i]
    component_matrix = np.outer(u, v)
    component_matrices[i] = component_matrix
    
    
vmin = min(np.min(e) for e in component_matrices)
vmax = max(np.max(e) for e in component_matrices)    
```

```python
fig, axes = plt.subplots(3, 2, figsize=(10, 15))
fig.suptitle('Component Matrices for Singular Value Decomposition of A', fontsize=16, y=1)

for i, ax in enumerate(axes.flatten()):       
    s = Sdiag[i]
    ax.set_title(f"Component Matrix for S = %.2f" % s)
    cm = component_matrices[i]
    im = ax.imshow(cm, cmap='viridis', vmin=vmin, vmax=vmax)  
    fig.colorbar(im, ax=ax) 
    

plt.tight_layout()
plt.show()
```

```python
A_alternate = np.sum(component_matrices * Sdiag[:, np.newaxis, np.newaxis], axis=0)
```

```python
rel_error = (A - A_alternate) / A

fig = plt.figure()
plt.title(r"Relative Error Between $A$ and $\sum_\alpha s_\alpha u_\alpha v_\alpha^T$", y=1.1)

plt.imshow(rel_error, cmap='viridis')  
plt.colorbar() 
```

## c)

```python
def MeanError(N):
    truncated_matrices = component_matrices[:N]
    trunctated_S = Sdiag[:N]
    A_alternate = np.sum(truncated_matrices * trunctated_S[:, np.newaxis, np.newaxis], axis=0)
    error = abs(A - A_alternate)
    mean_error = np.mean(error)
    return mean_error
```

```python
N_values = np.linspace(1, r, r, dtype=np.int32)
mean_errors = np.zeros(r) 

for i in range(len(N_values)):
    mean_errors[i] = MeanError(N_values[i])
```

```python
fig = plt.figure()
plt.plot(N_values, mean_errors, '.')
plt.title("Mean Error Between Origanal and Reconstructed Matrices")
plt.xlabel("Rank of Reconstruction")
plt.ylabel("Absolute Mean Error")
```

```python
def RadialVelocity(t, P, x):
    # Calculates the radial velocity of a star orbited by a planet
    # at the times in the vector t
    
    # extract the orbit parameters
    # P, t and tp in days, mp in Jupiter masses, v0 in m/s  
    mp, e, omega, tp, v0 = x
        
    # mean anomaly
    M = 2*np.pi * (t-tp) / P
    
    # velocity amplitude
    K = 204 * P**(-1/3) * mp  / np.sqrt(1.0-e*e) # m/s
    
    # solve Kepler's equation for the eccentric anomaly E - e * np.sin(E) = M
    # Iterative method from Heintz DW, "Double stars", Reidel, 1978
    # first guess
    E = M + e*np.sin(M)  + ((e**2)*np.sin(2*M)/2)
    while True:
        E0 = E 
        M0 = E0 - e*np.sin(E0)
        E = E0 + (M-M0)/(1.0 - e*np.cos(E0))
        if np.max(np.abs((E-E0))) < 1e-6:
            break
        
    # evaluate the velocities
    theta = 2.0 * np.arctan( np.sqrt((1+e)/(1-e)) * np.tan(E/2))
    vel = v0 + K * ( np.cos(theta + omega) + e * np.cos(omega))
    
    return vel
```

```python
def PartialDerivatives(t, P, x):
    # Calculates the numerical derivatives of the radial velocity with respect to the parameters
    y = RadialVelocity(t, P, x)
    A =  np.zeros([len(t), len(x)])
    eps = 1e-8
   
    for i in range(len(x)):
        x1 = np.copy(x)
        x1[i] = x[i] * (1 + eps)
        y1 = RadialVelocity(t, P, x1)
        pd = (y1 - y) / (x1[i]-x[i])
        A[:, i] = pd
    
    return A
```

```python
tobs, vobs, eobs = np.loadtxt('rvs.txt', unpack=True)

P = 1724
x = np.array([1.0, 0.5, 0.5, 0.5, 0.5])
```

```python
chisq = 1e99
lam = 1e-3
alpha = 0.2

while chisq > 1e-6:
    # compute the update to the parameters
    v = RadialVelocity(tobs, P, x)
    A = PartialDerivatives(tobs, P, x)
    r = (vobs-v)

    chisq = np.sum(r**2)
    lhs = A.T@A
    lhs = lhs@(np.identity(len(x))*(1+lam))
    rhs = A.T@r
    dx = np.linalg.inv(lhs)@rhs
    
    # calculate the chisq associated with the new parameters
    x1 = x + alpha * dx
    v1 = RadialVelocity(tobs, P, x1)
    A1 = PartialDerivatives(tobs, P, x1)
    r1 = vobs-v1
    chisq1 = np.sum(r1**2)
    print("chisq = %lg, mp = %lg, e = %lg, omega = %lg, tp = %lg, v0 = %lg, lam = %lg" % (chisq1 ,x[0], x[1], x[2], x[3], x[4], lam))
    
    # accept if chisq decreases; reject if it increases
    if chisq1 > chisq:
        lam = lam * 10
    else:
        lam = lam / 10
        x = x1
        # if the improvement in chisq becomes too small then exit
        if chisq-chisq1 < 1e-3:
            break
```
