# Question 1

## a)

The error associated with the Trapizoidal rule stems from the fact that this rule makes a linear approximation estimate $f(x)$ between each point. The lowest order neglected terms are therefore quadratic terms. That is to say that in the Taylor expansion of $f(x)$, we include only the first two terms and the error is $\mathcal{O}(\Delta x)^2$.

The error associated with the Simpson's rule stems from the fact that this rule makes a quadratic approximation to estimate $f(x)$ between each point. Following the same logic as for the Trapizoidal rule, we would expect the lowest order neglected terms to be cubic. However, the cubic terms end up reducing due to the symmetry of the approximation. The error is therefore $\mathcal{O}(\Delta x)^4$.

## b)

Let c be some proportionality constant.
Let n be the exponential scaling of error with stepsize for a given rule.

$\epsilon_1 = c(\Delta x)^n$
$\epsilon_2 = c(2\Delta x)^n$

$I = I_1 + \epsilon_1 = I_2 + \epsilon_2$

$I_1 - I_2 = \epsilon_2 - \epsilon_1$

$I_1 - I_2 = c(2\Delta x)^n - c(\Delta x)^n = c(\Delta x)^n * (2^n - 1)$

$c(\Delta x)^n = \epsilon_1 = \frac{I_1 - I_2}{2^n - 1}$

For the Trapasoidal rule: n=2

$\epsilon_1 = \frac{I_1 - I_2}{2^2 - 1} = \frac{I_1 - I_2}{3}$

For the Simpson's rule: n=4:

$\epsilon_1 = \frac{I_1 - I_2}{2^4 - 1} = \frac{I_1 - I_2}{15}$


The above assumes that the scale factor, c, will be the same in case 1 and case 2. This is an estimation and may not be exactly true. 

## c)

```python
# down load data
data = np.loadtxt('hw2_data.txt')

# retrieve x and y from data columns
x = data[:, 0]  
y = data[:, 1] 
```

```python
# Define functions that integrate take x and y data, integrate numerically, and return integral with approximate error

def Trapezoidal(x1, y1):
    # calculate the differece between evenly space x values
    dx1 = np.diff(x1)[0]  
    
    # perfom numerical integration
    I_1 = dx1 * ((1/2)*(y1[0] + y1[-1]) + np.sum(y1[1:-1]))
    
    # splice x and y arrays to retrieve half of the original points
    x2 = x1[0::2]
    y2 = y1[0::2]
    
    # perfom numerical integration with less points
    dx2 = np.diff(x2)[0]
    I_2 = dx2 * ((1/2)*(y2[0] + y2[-1]) + np.sum(y2[1:-1]))

    # estimate error
    epsilon = (I_1 - I_2) / 3
    
    # return both integrals and error
    return I_1, I_2, epsilon


def Simpson(x1, y1):
    # Calculate the differece between evenly space x values
    dx1 = np.diff(x1)[0] 
    
    # perfom numerical integration
    I_1 = (dx1 / 3) * (y1[0] + y1[-1] + 4*np.sum(y1[1:-1:2]) + 2*np.sum(y1[2:-2:2])) 
    
    # splice x and y arrays to retrieve half of the original points
    x2 = x1[0::2]
    y2 = y1[0::2]
    
    # perfom numerical integration with less points
    dx2 = np.diff(x2)[0]
    I_2 = (dx2 / 3) * (y2[0] + y2[-1] + 4*np.sum(y2[1:-1:2]) + 2*np.sum(y2[2:-2:2])) 
    
    # estimate error
    epsilon = (I_1 - I_2) / 15
    
    # return both integrals and error
    return I_1, I_2, epsilon
```

```python
# unpack integration functions
I1_t, I2_t, epsilon_t  = Trapezoidal(x, y)
I1_s, I2_s, epsilon_s  = Simpson(x, y)
```

```python
print('The approximate error using the Trapazoidal method is', epsilon_t)
print("The approximate error using the Simpson's method is", epsilon_s)
```

I find that the error with the Trapazoidal method is approximately 2 orders of magnitude larger than that with the Simpson's method. This is too be expected since the Simpson's method performs a higher order approximation to achieve better accuracy. Knowing that $\epsilon_T ~ (\Delta x)^2$ and $\epsilon_S ~ (\Delta x)^4$, we realize that $\sqrt{\epsilon_T}$ and $\sqrt[4]{\epsilon_S}$ should be on the same order of magnitude. Indeed

$\sqrt{\epsilon_T} = 0.02481507807861229$ and $\sqrt[4]{\epsilon_S} = 0.04574738333962722$

## a)

Performing 2 change of variables on the integral, we can simplify it to obtain $\frac{N}{Vn_Q}$ as a definite integral in x, dependant only on $\frac{\mu}{kT}$.

$N = V \int_0^\infty {8\pi p^2 dp\over h^3} {1\over 1+e^{(\epsilon-\mu)/kT}} = V \int_0^\infty {8\pi p^2 dp\over (2\pi h)^3} {1\over 1+e^{(\epsilon-\mu)/kT}} = V \int_0^\infty {p^2 dp\over \pi^2 h^3} {1\over 1+e^{(\epsilon-\mu)/kT}}$

$\rho = \sqrt{2m\epsilon} \rightarrow d\rho \sqrt{m\over 2\epsilon}d\epsilon$

$N = V \int_0^\infty {2m\epsilon \over \pi^2 h^3}\sqrt{m\over 2\epsilon} d\epsilon {1\over 1+e^{(\epsilon-\mu)/kT}}$

$N = V \int_0^\infty {\sqrt{2}\over \pi^2}{m^{3\over2}\over h^3} {\sqrt{\epsilon}d\epsilon \over 1+e^{(\epsilon-\mu)/kT}}$

$\epsilon = kTx \rightarrow d\epsilon = kTdx$

$N = V \int_0^\infty {\sqrt{2}\over \pi^2}({mkT\over h^2})^{3\over2} {\sqrt{x}dx \over 1+e^{x-(\mu/kT)}}$

$N = {4\over \sqrt{\pi}} V \int_0^\infty ({mkT\over 2\pi h^2})^{3\over2} {\sqrt{x}dx \over 1+e^{x-(\mu/kT)}}$

$N = {4\over \sqrt{\pi}}Vn_Q \int_0^\infty{\sqrt{x}dx \over 1+e^{x-(\mu/kT)}}$

${N\over Vn_Q} = {n\over n_Q} = {4\over \sqrt{\pi}} \int_0^\infty{\sqrt{x}dx \over 1+e^{x-(\mu/kT)}}$

```python
# define the integrand of the above integral making the following subsitution
# u = mu / kT

def Integrand(x, u):
    return (4 / np.sqrt(np.pi)) * np.sqrt(x) / (1 + np.exp(x - u))
```

```python
# define a range of mu / kT to solve at (from negative to positive)
u_values = np.linspace(-100, 100, 1000)

# define empty arrays to fill with the integral and associated error for each u
integral_values = np.zeros_like(u_values)
error_values = np.zeros_like(u_values)

# integrate for all u values
for i in range(len(u_values)):
    integral_values[i], error_values[i] = quad(Integrand, 0, np.inf, args=(u_values[i]))
```

```python
# interpolate over u_values and integral values 
# to obtain interpolation function that gives mu / kT as a function of n / n_Q
interp_u = CubicSpline(integral_values, u_values)
```

```python
# thin out u_values and integral_values to plot overtop of interpolation
u_points = u_values[::100]
I_points = integral_values[::100]

# plot interpolation with select p
x = np.logspace(-43, 3.3, 100)
plt.plot(x, interp_u(x), label='Interpolated Function')
plt.plot(I_points, u_points, '.', label='Numerically Integrated Points')
plt.xlabel('${n}/{n_Q}$')
plt.ylabel('${\mu}/{kT}$')
plt.xscale('log')
plt.legend()
```

## b)

There are two sources of error in above calculation. First, there is the error associated with the numerical integration necessary to obtain ${n\over n_Q}$ for a given value of ${\mu \over kT}$.  Then there is the error associated with interpolating over these arrays. The integration error at each point is returned by the funcition scipy.integrate.quad. The interpolation error is more difficult to approximate. 

```python
# define array of fractional error from integration
err_integrate = error_values / integral_values # error in n/n_Q
```

```python
# plot fractional error

F = plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title('Integration Error in $n/n_Q$')
plt.plot(u_values, err_integrate,)
plt.ylabel('Integration Error')
plt.xlabel('$\mu/kT$')

plt.subplot(1, 2, 2)
plt.title('Interpolation Error in $\mu/kT$')
plt.plot(integral_values, err_interpolate,)
plt.ylabel('Integration Error')
plt.xlabel('$n/n_Q$')          
```

## c)

The the nondegenerate and degenerate form can be rearranged such that they give ${\mu \over kT}$ as a function of ${n\over n_Q}$.

Nondegenerate:
$\mu = kT ln({n\over 2 n_Q})$

${\mu \over kT} = ln({1\over 2}{n\over n_Q})$

Degenerate:
$\mu = {\hbar^2 (3\pi^2n)^\frac{2}{3} \over 2m}$

${\mu \over kT} = {1\over 4}(9\pi)^\frac{1}{3}({n \over n_Q})^\frac{2}{3}$

```python
# Define these functions

def NonDegenerate(n_nQ):
    return(np.log(n_nQ / 2))

def Degenerate(n_nQ):
    return (1/4) * np.cbrt(9 * np.pi) * (n_nQ)**(2/3)
```

```python
# Plot with interpolated function to see overlap

x = np.logspace(-43, 3.3, 1000)

plt.plot(x, NonDegenerate(x), label='Nondegenerate limit')
plt.plot(x, Degenerate(x), label='Degenerate limit')
plt.plot(x, interp_u(x), '--', label ='Interpolated Function')

plt.xlabel('${n}/{n_Q}$')
plt.ylabel('${\mu}/{kT}$')
plt.xscale('log')
plt.legend()
```

```python
# calculate the fractional error between the interpolated values of mu / kT and the values for each limiting case
non_degenerate_error = abs((interp_u(x) - NonDegenerate(x))/ interp_u(x))
degenerate_error = abs((interp_u(x) - Degenerate(x))/ interp_u(x))
```

```python
# determine bounds in n/n_Q for which the percent error of each limiting case is less than 1%
non_degenerate_space = x[non_degenerate_error <= 0.01]
degenerate_space = x[degenerate_error <= 0.01]

print('The non degenerate limit is a valid approximation for values of n/n_Q less than', non_degenerate_space[-1])
print('The degenerate limit is a valid approximation for values of n/n_Q greater than', degenerate_space[0])
```

## a)

The Maxwell Boltzmann distribution is as follows;

$f(v) = 4\pi v^2 ({m \over 2\pi kT})^{3/2} e^{- m v^2\over 2kT}$

Making the subsitution $\sigma^2 = {kT\over m} gives

$f(v) = 4\pi v^2 ({1 \over 2\pi \sigma^2})^{3/2} e^{- v^2\over 2\sigma^2}$

Setting $\sigma = 0$ gives

$f(v) = \sqrt{2\over \pi} e^{- v^2\over 2}$

```python
# Define this function
def fct(v):
    return np.sqrt(2 / np.pi) * v**2 * np.exp(-v**2 / 2)
```

```python
# define the number generator using the rejection method
def MaxwellBoltzmann(N): 
    N = int(N)
    
    seed = 100    
    rng = np.random.default_rng()
    
    # Generate 4 time N random points in the x-y grid
    x = rng.uniform(0, 5, 4*N)
    y = rng.uniform(0, 0.6, 4*N)
    
    # define maximum y value for points to be in distribution
    y_limit = fct(x)

    # reject points outside curve
    x_rejected = x[y>y_limit]
    y_rejected = y[y>y_limit]

    x = x[y<=y_limit]
    y = y[y<=y_limit]
    
    return x[:N], y[:N], x_rejected[:N], y_rejected[:N]
```

```python
# Test that the function can generate 1 millions points in less than 1 second

t0 = time.time()

x, y, x_rejected, y_rejected = MaxwellBoltzmann(1e6)

t1 = time.time()
print('That took', t1-t0, ' seconds')
```

```python
# Plot generated point, rejected points, and curve

xx = np.linspace(0, 5, 100)
plt.plot(x_rejected, y_rejected, '.', ms=0.1, label='blue: Rejected Points')
plt.plot(x, y, '.', ms=0.1, label='orange: Generated Points')
plt.plot(xx, fct(xx), 'k')
plt.ylabel('$f(v)$')
plt.xlabel('$v$')
plt.legend()
```

This generator works by generating 4 times the desired number of points using a uniform random number generator over the x-y grid. It then eliminates anypoints for which the y value does not fall beneath the desired curve. Finally it trims the arrays to N points. Since the distribution takes up roughly one third of the grid, there are alway at least N points left after rejection.

```python
# define function that returns fractional error associate with a generating a given number of points

def fractional_error(N):
    
    # define analytical mean subsituting in sigma=1
    analytical_mean = np.sqrt(8 / np.pi)
    
    x = MaxwellBoltzmann(N)[0]
    
    # Calculate numerical mean
    numerical_mean = np.mean(x)
    
    # Calculate and return error
    error = abs(analytical_mean - numerical_mean)/analytical_mean
    return error
```

```python
# calculate error for a range of N values
N_values = np.linspace(1, 1e6, 100)
Error = np.zeros_like(N_values)
for i in range(len(N_values)):
    Error[i] = fractional_error(N_values[i])
```

```python
# Plot error

plt.plot(N_values, Error)
plt.xlabel('Number of Points generated')
plt.ylabel("Fractional error in the mean velocity")
```
