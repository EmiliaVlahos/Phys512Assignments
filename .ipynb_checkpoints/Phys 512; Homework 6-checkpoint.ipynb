## 1. Pseudo-Hamiltonian for the simple harmonic oscillator

## a)


For some position and velociy $x_n$ and $v_n$, the Pseudo-Hamiltonian is:

$\tilde{H}_n(h) = {1\over 2}x_n^2 + {1\over 2} v_n^2 {1\over 1-h^2/4}$.

Using the Jacobian for the leapfrog method to update position and velocity, we find

$x_{n+1} = (1 - {h^2\over 2})x_n + hv_n$,

$v_{n+1} = ({h^3\over 4} - h)x_n + (1 - {h^2\over 2})v_n$.

From this, we can calculate the Pseudo-Hamiltonian at the updated timestep:

$x_{n+1}^2 = (1 - {h^2\over 2})^2x_n^2 + 2h(1 - {h^2\over 2})x_nv_n + h^2v_n^2$

$x_{n+1}^2 = (1 - h^2 + {h^4\over 4})x_n^2 + (2h - h^3)x_nv_n + h^2v_n^2$

$v_{n+1}^2 = ({h^3\over 4} - h)^2x_n^2 + 2({h^3\over 4} - h)(1 - {h^2\over 2})x_nv_n + (1 - {h^2\over 2})^2v_n^2$

$v_{n+1}^2 = h^2(1 - {h^2 \over 4})^2x_n^2 - 2h(1 - {h^2\over 4})(1 - {h^2\over 2})x_nv_n + (1 - h^2 + {h^4\over 4})v_n^2$

${v_{n+1}^2 \over (1 - {h^2\over 4})} = (h^2 - {h^4 \over 4})x_n^2 - (2h - h^3)x_nv_n + {(1 - h^2 + {h^4\over 4})\over(1 - {h^2\over 4})}v_n^2$

$\tilde{H}_{n+1}(h) = {1\over 2}(x^2_{n+1}) + {1\over 2} (v^2_{n+1}) {1\over 1-h^2/4}$

$\tilde{H}_{n+1}(h) = {1\over 2}[x^2_{n+1} + {v^2_{n+1}\over 1-h^2/4}]$

$\tilde{H}_{n+1}(h) = {1\over 2}[(1 - h^2 + {h^4\over 4})x_n^2 + (2h - h^3)x_nv_n + h^2v_n^2 + (h^2 - {h^4 \over 4})x_n^2 - (2h - h^3)x_nv_n + {(1 - h^2 + {h^4\over 4})\over(1 - {h^2\over 4})}v_n^2]$

$\tilde{H}_{n+1}(h) = {1\over 2}[x_n^2 + (h^2 + {(1 - h^2 + {h^4\over 4})\over(1 - {h^2\over 4})})v_n^2]$

$\tilde{H}_{n+1}(h) = {1\over 2}[x_n^2 + v_n^2{1\over(1 - {h^2\over 4})}]$

$\tilde{H}_{n+1}(h) = {1\over 2}x_n^2 + {1\over 2} v_n^2 {1\over 1-h^2/4} = \tilde{H}_n(h)$

We conclude that the pseudo hamilitonian is conserved in the leapfrog method.


## b)

```python
def leapfrog(nsteps, dt, IC):
    '''
    Implements the leapfrog method to solve for the position and velocity of a harmonic oscillator
    
    Arguments:
    - nsteps (int): Number of steps
    - dt (float): Step size
    - IC (array): x and y coordinate at time 0
    
    Returns:
    - xv.T (array): 2 by nsteps+1 array containing positions (first row) and velocities (second row) for each timestep
    '''
    
    # Create array to record position and velocity updates and set zeroeth position to ICs
    xv = np.zeros((nsteps+1, 2))
    xv[0] = IC
    
    # Define the jacobean for the leapfrop update of a harmonic oscillator
    J = np.array([[(1-dt**2/2), dt], [(dt**3/4-dt), (1-dt**2/2)]])
    
    # Iterate through nsteps and record x and v at each step
    for i in range(1, nsteps+1):
        xv[i] = J@xv[i-1]
        
    return xv.T
```

```python
def PseudoHamiltonian(x, v):
    '''
    Calculates the pseudo hamiltonian for a given velociy and position
    '''
    H = 1/2 * (x**2 + v**2 / (1 - dt**2/4))
    return H
```

```python
# Define parameters to implement leapfrog method over single oscillation
nsteps = 1000
tf = 2*np.pi
dt = tf/nsteps
t = dt * np.arange(nsteps+1)

x0 = 1
v0 = 0
IC = np.array([x0, v0])
```

```python
# Run leapfrog 
x, v = leapfrog(nsteps, dt, IC)
```

```python
# Plot position and velocity solutions
plt.title("Harmonic Oscillator Solved via Leapfrog Method")
plt.plot(t, x, label='Position', color='pink')
plt.plot(t, v, label='Velocity', color='violet')
plt.xlabel('Time')
plt.legend()
```

```python
# Calculate the Pseudo Hamiltonian at each timestep
H = PseudoHamiltonian(x, v)

# Plot the error in the Pseudo Hamiltonian
plt.title('Error in the Pseudo Hamiltonian')
plt.plot(t, abs(H-H[0]))
plt.yscale('log')
plt.xlabel('Time')
plt.ylabel('log Error')

print("The error in the Hamiltonian after 1000 timesteps is:", H[-1]-H[0])
```

From the above plot we see that the Pseudo Hamiltonian does seem to increase with each step. That is to say it is not perfectly conserved. However the extent of this increase is very small. After integrating over 1 period of oscillation we have an error of 1e-14. This is likely the accumulation of round off error at each step.

## c)

```python
def Energy(x, v):
    '''
    Calculates the energy of the system for a given velociy and position
    '''
    E = 1/2 * (x**2 + v**2)
    return(E)
```

```python
# Calculate the energy at each timestep and plot the error 
E = Energy(x, v)

plt.title("Error in Total Energy")
plt.plot(t, E-E[0])
plt.xlabel("Time")
plt.ylabel("Error")
```

The plot above demontrates how the energy error associated with the leapfrog method relates to the fraction of an oscillation completed. To further investigate this we will observe how error scales with number of steps at $t=\pi$ and at $t={\pi \over 2}$.

```python
# define a range of nsteps and calculate the associated timesteps
n = np.logspace(1, 4, 1000, dtype=int)
tf = np.pi
dt = tf/n

# Initialize arrays to track error at pi/2 (half) and at pi (full)
E_half = np.zeros(len(n))
E_full = np.zeros(len(n))

# iterate through the different values of nsteps, run the leapfrog method, and calculate error at pi/2 and at pi
for i in range(len(n)):
    x, y = leapfrog(n[i], dt[i], IC)
    E = Energy(x, y)
    E_half[i] = E[len(E) // 2] - E[0]
    E_full[i] = E[-1] - E[0]
```

```python
nscaling = n[:700]

# plot error
plt.title("Error in Total Energy")

plt.plot(n, abs(E_half), color='dodgerblue', label='After half cycle')
plt.plot(nscaling, 1/nscaling**2, '--', color='deeppink', label='$1/n^2$')
plt.plot(n, abs(E_full), color='crimson', label='After full cycle')
plt.plot(nscaling, 1/nscaling**6, '--', color='mediumorchid', label='$1/n^6$')
plt.xlabel("Number of steps per cycle (of period $\pi$)")
plt.ylabel("Error")
plt.xscale('log')
plt.yscale('log')
plt.legend()
```

The analitice expression for E-E_0 is 

$E-E_0 = {1\over 16} dt^6 \cdot x^2 - {1\over 4} dt^5 \cdot xv + {1\over 4} dt^4 \cdot (v^2 - x^2) + {1\over2} dt^3 \cdot xv$.

Notice that for $x = cos(t)$ and $v = -sin(t)$, the lower order terms $dt^5, dt^4$, and $dt^3$ contain expressions of x and v that average to zero over the interval 0 to $\pi$. This means that the error from these terms will cancel after a full cycle. We are left with an error that scales with $dt^6$ or alternatively $1\over n^6$. This is what we observe :). For a half cycle, we do not expect to see this cancelation. This leaves us with an error that scales with  $dt^3$ or alternatively $1\over n^3$. However, we instead observe an error that scales with  $1\over n^2$. Thats kind of weird. 

## 2. Eigenvalue problem for the wave on a string

## a)

#### The PDE for a wave on a string is: 
$$ \rho (x) \frac{\partial^2f}{\partial t^2} - T\frac{\partial^2f}{\partial x^2} = 0 $$

#### For the normal mode of the wave, $\xi\propto e^{i\omega t}$. The time derivative therefore reduces to:
$$ \rho (x) \frac{\partial^2f}{\partial t^2} = - \rho (x) \omega^2 f$$

#### Our wave equation thus becomes:
$$ - \rho (x) \omega^2 f - T\frac{\partial^2f}{\partial x^2} = 0 $$

$$ \frac{\partial^2f}{\partial x^2} = -\frac{\omega^2\rho(x)}{T}f$$

#### We will set T=1 everywhere in the string for simplicity. Finite differencing this over a grid x, we obtain
$$  {f_{i+1}-2f_i+f_{i-1}\over (\Delta x)^2} = - \omega^2\rho_i f_i, $$

#### Here $f$ is a vector describing the hieght of the string wave at each index of $x$. $\rho_i$ is the density at a given point in on x. We can represent the above equation for all i values using matrix multiplication. We assume clamped boundary conditions such that $f_0 = f_{n+1} = 0$ and we exclude these points from the vector $f$.

\begin{array}{c|c}
\frac{1}{(\Delta x)^2}
\begin{bmatrix} 
2 & -1 & 0 & \dots & 0 \\
-1 & 2 & -1 & \dots & 0 \\
0 & -1 & 2  & \dots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & -1 & 2 \\
\end{bmatrix} 
\begin{bmatrix}
    f_1 \\
    f_2 \\
    f_3 \\
    \vdots \\
    f_n \\
\end{bmatrix}\hspace{1cm} = \hspace{1cm}
\omega^2
\begin{bmatrix} 
\rho_1 & 0 & 0 & \dots & 0 \\
0 & \rho_2 & 0 & \dots & 0 \\
0 & 0 & \rho_3  & \dots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 &  & \rho_n \\
\end{bmatrix} 
\begin{bmatrix}
    f_1 \\
    f_2 \\
    f_3 \\
    \vdots \\
    f_n \\
\end{bmatrix}\hspace{1cm}
\end{array}

#### The above equation is of the form
$$\mathbf{A}\cdot\mathbf{f} = \omega^2 \mathbf{b}\cdot\mathbf{f}$$
#### Where A is the tridiagonal matrix describing the finite differenced differential equation and b is the tridiagonal matrix describing the density of the string along the grid. 

## b)

```python
# Choose a string of length 1 and divide grid into 1000 segments of width dx
N = 1000
L = 1
dx = L/N
x = dx * np.arange(N+1)
```

```python
# Define the tridiagonal matrix A 
A = np.zeros((N-1, N-1))
for i in range(N-1):
    A[i][i] = 2   
    if i >= 1:
        A[i][i - 1] = -1
    if i < N-2:
        A[i][i + 1] = -1
A /= dx**2

# Define the diagonal matrix b, with each entry correspoinding to the density at the related position x
b = np.identity(N-1)

# Solve for the eigan values and vectors
w, f = eigh(A, b)

# Append the clamped boundary conditions to the first and last position of each eigan vector
first_last = np.zeros((1, len(w)))
f = np.vstack((first_last, f, first_last))
```

```python
# plot the first 8 eigan functions

fig, axes = plt.subplots(8, 1, figsize = (10,20))
fig.suptitle('Normal Modes for Wave on String with Constant Density', fontsize=25, y=1.01)

for i, ax in enumerate(axes.flatten()):    
    omega = np.sqrt(w[i])/np.pi 
    ax.plot(x, f[:, i])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f"Angular Frequency = %.5f pi" % omega)
    
    
plt.tight_layout()
plt.show()
```

I find that for a string with constant density, the eigan frequencies come in units of $n\pi$ for n=1, 2, 3, 4, 5, 6, 7, 8, etc. This is also what we found using the shooting method. Whereas the shooting method seemed to overestimate frequency by a very small percent, our method underestimates frequency by a very small percent. 

```python
# Define rho(x) for a non constant density, assuming that the scale factor is 1
rho = 1 + 10*x**2

# Use rho(x) to define new diagonal matrix b
rho = rho[1:-1]
b = np.diag(rho)

# Solve for the eigan values and vectors
w, f = eigh(A, b)

# Append the clamped boundary conditions to the first and last position of each eigan vector
first_last = np.zeros((1, len(w)))
f = np.vstack((first_last, f, first_last))
```

```python
# plot the first 8 eigan functions

fig, axes = plt.subplots(8, 1, figsize = (10,20))
fig.suptitle('Normal Modes for Wave on String with Non-Constant Density', fontsize=25, y=1.01)

for i, ax in enumerate(axes.flatten()):    
    omega = np.sqrt(w[i])/np.pi 
    ax.plot(x, f[:, i])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f"Angular Frequency = %.5f pi" % omega)
    
    
plt.tight_layout()
plt.show()
```

The eigan frequencies are once again the same as what we calculated using the shooting method. 

## 3. Leapfrogging an electromagnetic wave

## a)

```python
def leapfrog(Nsteps, dt, N, dx, B0, E0, ref_index):
    '''
    Implements the leapfrog method to solve for the propagation of electric and magnetic fields undergoing mutual induction
    
    Arguments:
    - Nsteps (int): Number of steps
    - dt (float): Step size
    - N (int): Number grid divisions
    - dx (float): Width of grid division
    - B0 (array): Magnetic field at time 0
    - E0 (array): Electric field at time 0
    - ref_index (array): Refractive index corresponding to grid position
    
    Returns:
    - B (array): Magnetic field at time for each timestep
    - E (array): Electric field at time for each timestep
    '''
    
    # Creates arrays to record B an E fields and set zeroeth position to initial condition
    B = np.zeros((Nsteps, N))
    E = np.zeros((Nsteps, N))
    
    B[0] = B0
    E[0] = E0
    
    # Propagate B and E fields for first timestep
    B[1] = B[0] - (dt/dx) * (np.roll(E[0], -1) - np.roll(E[0], 1)) 
    E[1] = E[0] - (dt/dx) * (np.roll(B[0], -1) - np.roll(B[0], 1)) / ref_index**2
    
    # iterate through the rest of the timesteps and propagate B and E fields
    for n in range(1, Nsteps-1):
        B[n+1] = B[n-1] - (dt/dx) * (np.roll(E[n], -1) - np.roll(E[n], 1)) 
        E[n+1] = E[n-1] - (dt/dx) * (np.roll(B[n], -1) - np.roll(B[n], 1)) / ref_index**2
    
    return B, E
```

```python
# Define a grid x of length L divided into N segments
N = 1000
L = 1
dx = L/N
x = dx * np.arange(N)

# Define a time range t of length tf divided into Nsteps timesteps
Nsteps = 5000
tf = 1
dt = tf/Nsteps
t = dt * np.arange(Nsteps)

# set the refractive index to 1 everywhere on the grid
ref_index = np.ones(N)
```

```python
# Initiate a the B and E fiels as gaussian profiles centered in the grid
mu = L/2   
sigma = 0.05

E0 = np.exp(-0.5 * ((x - mu) / sigma) ** 2)
B0 = np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Call the leapfrog function to propogate these fields
B, E = leapfrog(Nsteps, dt, N, dx, B0, E0, ref_index)
```

```python
%matplotlib 

# Initialize the figure and plot the first frame
plt.figure()
plt.title("Electric And Magnetic Feilds Propagating From Gaussian")
plt.plot(x, B0, label='B Feild')
plt.plot(x, E0, label='E Feild')
plt.xlabel("Grid")
plt.ylabel("Feilds")
plt.legend()
plt.show()

# Iterate through all timesteps and plot in increments of 10
for i in range(len(B)):
    
    if i % 10 == 0:
        plt.clf()
        plt.title("Electric And Magnetic Feilds Propagating From Gaussian")
        plt.plot(x, B[i], label='B Feild')
        plt.plot(x, E[i], label='E Feild')
        plt.xlabel("Grid")
        plt.ylabel("Feilds")
        plt.legend()
        plt.show()
        plt.pause(1e-10)
    
# Plot the last frame
%matplotlib inline
plt.clf()
plt.title("Electric And Magnetic Feilds Propagating From Gaussian")
plt.plot(x, B[-1], label='B Feild')
plt.plot(x, E[-1], label='E Feild')
plt.xlabel("Grid")
plt.ylabel("Feilds")
plt.legend()
plt.show()
```

```python
# Initiate a the B and E fiels as oppositely oriented gaussian profiles centered in the grid
E0 = -np.exp(-0.5 * ((x - mu) / sigma) ** 2)
B0 = np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Call the leapfrog function to propogate these fields
B, E = leapfrog(Nsteps, dt, N, dx, B0, E0, ref_index)
```

```python
%matplotlib 

# Initialize the figure and plot the first frame
plt.figure()
plt.title("Electric And Magnetic Feilds Propagating From Gaussian")
plt.plot(x, B0, label='B Feild')
plt.plot(x, E0, label='E Feild')
plt.xlabel("Grid")
plt.ylabel("Feilds")
plt.legend()
plt.show()

# Iterate through all timesteps and plot in increments of 10
for i in range(len(B)):
    
    if i % 10 == 0:
        plt.clf()
        plt.title("Electric And Magnetic Feilds Propagating From Gaussian")
        plt.plot(x, B[i], label='B Feild')
        plt.plot(x, E[i], label='E Feild')
        plt.xlabel("Grid")
        plt.ylabel("Feilds")
        plt.legend()
        plt.show()
        plt.pause(1e-10)
    
# Plot the last frame
%matplotlib inline
plt.clf()
plt.title("Electric And Magnetic Feilds Propagating From Gaussian")
plt.plot(x, B[-1], label='B Feild')
plt.plot(x, E[-1], label='E Feild')
plt.xlabel("Grid")
plt.ylabel("Feilds")
plt.legend()
plt.show()
```

The gaussian fields propagate without loosing their shape. When the B and E fields are both oriented upwards, they propagate forwards at the same velocity. When the B and E fields are oriented in opposite directions, they propagate backwards in time at the same velocity. 

## b)

```python
# Define a grid x of length L divided into N segments
N = 1000
L = 1
dx = L/N
x = dx * np.arange(N)

# Define a time range t of length tf divided into Nsteps timesteps
Nsteps = 5000
tf = 0.6
dt = tf/Nsteps
t = dt * np.arange(Nsteps)

# set the refractive index to 1 on the LHS of the grid and to 2 on the RHS of the grid
n1 = 1
n2 = 2

ref_index = np.ones(N)

ref_index[:int(N/2)] = n1
ref_index[int(N/2):] = n2
```

```python
# Initialize the B and E fiels as gaussian profiles on LHS of the grid
mu = L/5
sigma = 0.05

E0 = np.exp(-0.5 * ((x - mu) / sigma) ** 2) / n1
B0 = np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Call the leapfrog function to propogate these fields
B, E = leapfrog(Nsteps, dt, N, dx, B0, E0, ref_index)
```

```python
%matplotlib 

# Initialize the figure and plot the first frame
plt.figure()
plt.title("Electric And Magnetic Fields Propagating From Gaussian")
plt.plot(x, B0, label='B Field')
plt.plot(x, E0, label='E Field')
plt.axvline(L/2, color='gray')
plt.xlabel("Grid")
plt.ylabel("Feilds")
plt.legend()
plt.show()

# Iterate through all timesteps and plot in increments of 10
for i in range(len(B)):
    
    if i % 10 == 0:
        plt.clf()
        plt.title("Electric And Magnetic Fields Propagating From Gaussian")
        plt.plot(x, B[i], label='B Field')
        plt.plot(x, E[i], label='E Field')
        plt.axvline(L/2, color='gray')
        plt.xlabel("Grid")
        plt.ylabel("Fields")
        plt.legend()
        plt.show()
        plt.pause(1e-10)
    
# Plot the last frame
%matplotlib inline
plt.clf()
plt.title("Electric And Magnetic Fields Propagating From Gaussian")
plt.plot(x, B[-1], label='B Field')
plt.plot(x, E[-1], label='E Field')
plt.axvline(L/2, color='gray')
plt.xlabel("Grid")
plt.ylabel("Fields")
plt.legend()
plt.show()
```

The wave propagates forward until it encounters the barier where n changes from 1 to 2. Here, a portion of it is reflected and a portion is transmitted. The fraction of the amplitude that is reflected is given by of the reflected portion is given by $n_2-n_1\over n_1 + n_2$ = $1\over3$. This is what we observe. Note that the electric feild becomes inverted. Infact it scales like $n_1-n_2\over n_1 + n_2$ = $-1\over3$. This is necessary for the reflected waveform to propagate backwards, as seen in a). The fraction of the amplitude that is transmitted for the B feild is given by $2n_2\over n_1 + n_2$ = $4\over3$. The fraction of the amplitude that is transmitted for the E feild is given by $2n_1\over n_1 + n_2$ = $2\over3$. This ensures that on the RHS of the grid we have $n_2E = B$, allowing the waveform to propate forwards in one direction. The simulation therefore does display the expected behaviour as it crosses the boundary. Very cool!. (This was actually probably my favourite assigment question in this course. Super interesting)
